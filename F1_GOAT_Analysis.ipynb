```python
# Create a dataframe for each season containing driver and constructor standings
def get_standings_by_season(data):
    standings_by_season = {}
    for year in sorted(data['year'].unique()):
        season_data = data[data['year'] == year]
        
        # Driver standings
        driver_standings = season_data.groupby('driver')['points'].sum().sort_values(ascending=False).reset_index()
        driver_standings.columns = ['Driver', 'Total Points']
        
        # Constructor standings
        constructor_standings = season_data.groupby('constructor_name')['points'].sum().sort_values(ascending=False).reset_index()
        constructor_standings.columns = ['Constructor', 'Total Points']
        
        standings_by_season[year] = {
            'Driver Standings': driver_standings,
            'Constructor Standings': constructor_standings
        }
    return standings_by_season

get_standings_by_season(f1_data)
```

```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# Calculate championships won by each driver
def calculate_championships(standings_by_season):
    driver_championships = Counter()
    
    for year, standings in standings_by_season.items():
        # Get the driver standings for the year
        driver_standings = standings['Driver Standings']
        
        # The champion is the driver with the most points (top row)
        if not driver_standings.empty:
            champion = driver_standings.iloc[0]['Driver']
            driver_championships[champion] += 1

    # Convert Counter to a DataFrame for better display
    championships_df = pd.DataFrame(driver_championships.items(), columns=['Driver', 'Championships Won']).sort_values(by='Championships Won', ascending=False)
    
    return championships_df

# Calculate championships based on standings
championships_df = calculate_championships(get_standings_by_season(f1_data))

top_10_championships_df = championships_df.head(10)

plt.figure(figsize=(12, 6))
colors = plt.cm.tab10(range(len(top_10_championships_df)))  # Use tab10 colormap for top 10


plt.bar(top_10_championships_df['Driver'], top_10_championships_df['Championships Won'], color=colors)

plt.title('Top 10 Drivers by Championships Won', fontsize=16)
plt.xlabel('Driver', fontsize=12)
plt.ylabel('Championships Won', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Show the graph
plt.show()
```

6. However, the method of awarding points by finish has evolved over the years. Let's normalize the all-time points data to the current way of points scoring.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Add normalized points based on final racing position
def add_normalized_points(data):
    # Normalized points for each position(latest method)
    points_map = {
        1: 25,
        2: 18,
        3: 15,
        4: 12,
        5: 10,
        6: 8,
        7: 6,
        8: 4,
        9: 2,
        10: 1
    }
    
    # Assign normalized points based on final position
    data['normalized_points'] = data['final_position'].apply(lambda x: points_map.get(x, 0))
    return data

# Apply normalized points calculation to our dataset
f1_data = add_normalized_points(f1_data)
```

```python
import pandas as pd


# Calculate total points per constructor
constructor_points = f1_data.groupby('constructor_name')['normalized_points'].sum().sort_values(ascending=False)

# Assign ranks based on total points
constructor_ranks = constructor_points.rank(ascending=False, method='min').astype(int)


constructor_rankings = pd.DataFrame({
    'Constructor': constructor_points.index,
    'Total Points': constructor_points.values,
    'Rank': constructor_ranks.values
})

print(constructor_rankings)
```



We will weigh each driver's points by giving them weightage according to their constructor's historical rank. The logic would be that points earned with worse constructors are more valuable.



```python
# Merge constructor ranks back into the main dataset
f1_data = f1_data.merge(constructor_rankings[['Constructor', 'Rank']], left_on='constructor_name', right_on='Constructor')

# Calculate weighted points: higher rank (weaker constructor) gives more weight
f1_data['Weighted Points_1'] = f1_data['normalized_points'] *  f1_data['Rank']

driver_circuit_performance = f1_data.groupby(['driver'])['Weighted Points_1'].sum().reset_index()

# Sorting by 'Weighted Points' column in descending order
driver_circuit_performance = driver_circuit_performance.sort_values(by='Weighted Points_1', ascending=False)

```

8. Weight Driver Performances by Dynamic Constructor Strength

We realised that historically ranking the constructors gave bias to drivers with newer constructors like Red Bull or Brawn who only had one season. We realised that was why Jenson Button's numbers were inflated in the previous visualization. 

So, we decided to rank the constructors dynamically each year on the basis of their performance the previous year. 

This method would give value to drivers who are adaptable to new situations, and don't rely on a tried and tested car, or a successful formula of personnel, equipment and familiarity to rack up winning streaks. Ahem, Hamilton, ahem.

```python
def add_constructor_season_rank(data, standings):
    data['constructor_season_rank'] = None
    for year, season_data in standings.items():
        if year + 1 in data['year'].values:  # Check if next season exists
            constructor_standings = season_data['Constructor Standings']
            for index, row in constructor_standings.iterrows():
                constructor = row['Constructor']
                rank = index + 1
                data.loc[(data['year'] == year + 1) & (data['constructor_name'] == constructor), 'constructor_season_rank'] = rank
    return data


f1_data=add_constructor_season_rank(f1_data, get_standings_by_season(f1_data))
```

```python
# Ensure 'constructor_season_rank' is numeric
f1_data['constructor_season_rank'] = pd.to_numeric(f1_data['constructor_season_rank'], errors='coerce')

# Calculate weighted points: higher rank (weaker constructor) gives more weight
f1_data['Weighted Points_2'] = f1_data['normalized_points'] * f1_data['constructor_season_rank']

f1_data['Weighted Points_2'] = pd.to_numeric(f1_data['Weighted Points_2'], errors='coerce')

# Aggregate weighted points by driver
driver_circuit_performance = f1_data.groupby(['driver'])['Weighted Points_2'].sum().reset_index()


driver_circuit_performance['Weighted Points_2'] = pd.to_numeric(driver_circuit_performance['Weighted Points_2'], errors='coerce')

# Sorting by 'Weighted Points' column in descending order
driver_circuit_performance = driver_circuit_performance.sort_values(by='Weighted Points_2', ascending=False)

# Filter the top 10 drivers by Weighted Points
top_10_drivers = driver_circuit_performance.nlargest(10, 'Weighted Points_2')


plt.figure(figsize=(12, 6))
colors = plt.cm.tab10(range(len(top_10_drivers)))  # Use a vibrant colormap

plt.bar(top_10_drivers['driver'], top_10_drivers['Weighted Points_2'], color=colors)

plt.title('Top 10 Drivers by Weighted Points - Method 2', fontsize=16)
plt.xlabel('Driver', fontsize=12)
plt.ylabel('Weighted Points - 2', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

plt.show()
```
